```python
import pandas as pd
```

```python
#서울시 구별 건강지표
health_total = pd.read_csv("health_total.csv", encoding = "utf-8")
health_total.head()
```

```python
#구별을 구분으로 바꿈
health_total.rename(columns={health_total.columns[0] : '구분'}, inplace=True)
health_total.head()
```

```python
health_total.head()
```

```python
#서울시 가구원수별 가구형태 통계
single_household = pd.read_csv("number.csv", encoding = "utf-8", thousands = ',')
single_household.head()
```

```python
#서울시 구별 1인가구비율 구하기
single_household['1인가구비율'] = (single_household['1인가구'] / single_household['계']) * 100
single_household.head()
```

```python
#1인가구와 1인가구비율 외 나머지 열 삭제
del single_household['2인가구']
del single_household['3인가구']
del single_household['4인가구']
del single_household['5인 가구']
del single_household['6인가구']
del single_household['7인이상 가구']
single_household.head()
```

```python
single_household = single_household.iloc[1:,:]
```

```python
single_household.head()
```

```python
# '구분'을 중심으로 합치기
data_result = pd.merge(single_household, health_total, on='구분')
data_result.head()
```

```python
data_result.set_index('구분', inplace=True)
data_result.head()
```
