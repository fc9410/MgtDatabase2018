```python
# 그래프 그리기를 위한 set_index
household_gu.set_index("구별", inplace=True)
```

```python
plt.figure()
plt.title('1인가구비율')
household_gu['1인가구비율'].plot(kind='barh', grid=True, figsize=(10,10)) # 수평 막대 그래프를, 그리드를 사용하여, 그림 크기를 10*10으로 지정함.
plt.show()
```

```python
# sort_values 사용
household_gu['1인가구비율'].sort_values().plot(kind='barh', 
                                     grid=True, figsize=(10,10))
plt.show()
```

```python
import folium
```

```python
# 한국 지도 형태 가져오기
import json
geo_path = 'skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
```

```python
# 1인가구비율로 그리기
map = folium.Map(location=[37.5502, 126.982], zoom_start=10, 
                 tiles='Stamen Toner')

map.choropleth(geo_data = geo_str,
               data = household_gu['1인가구비율'],
               columns = [household_gu.index, household_gu['1인가구비율']],
               fill_color = 'PuRd', #PuRd, YlGnBu
               key_on = 'feature.id')
map
```
